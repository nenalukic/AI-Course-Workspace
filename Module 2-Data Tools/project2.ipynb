## Module 2 Mini-Project: Health Data EDA (Exploratory Data Analysis)
The Task:
You are a data scientist looking at patient health data. You will generate a mock dataset, clean a missing value, calculate the average heart rate, and plot the relationship between Age and Cholesterol levels.

## Cell 1: Setup & Data Creation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate mock patient data
patient_data = {
    "Patient_ID": range(1, 11),
    "Age": [25, 45, 30, 50, 65, 35, 40, np.nan, 55, 60], # np.nan is a true missing value
    "Heart_Rate": [70, 85, 72, 90, 88, 75, 82, 78, 92, 85],
    "Cholesterol": [180, 220, 190, 240, 260, 200, 210, 195, 250, 245]
}

df_health = pd.DataFrame(patient_data)
display(df_health.head()) # .head() shows the first 5 rows
```

## Cell 2: Data Cleaning and Analysis

```python
# 1. Fill the missing Age with the median age of the dataset
median_age = df_health["Age"].median()

# Assigning it directly instead of using inplace=True
df_health["Age"] = df_health["Age"].fillna(median_age)

# 2. Calculate the average heart rate
avg_hr = df_health["Heart_Rate"].mean()
print(f"The average heart rate across all patients is: {avg_hr} bpm")

# 3. Create a new column categorizing 'High Cholesterol' (over 200)
df_health["High_Cholesterol"] = df_health["Cholesterol"] > 200
display(df_health)
```

## Cell 3: Visualization

```python
# Plot Age vs Cholesterol, coloring the dots based on whether cholesterol is high
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_health, 
    x="Age", 
    y="Cholesterol", 
    hue="High_Cholesterol", 
    palette="Set1",
    s=150
)

plt.title("Patient Age vs. Cholesterol Levels")
plt.show()
```
