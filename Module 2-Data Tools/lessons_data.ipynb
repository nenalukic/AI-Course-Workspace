## Lesson 2: NumPy (Numerical Python)
NumPy is the backbone of almost all AI and machine learning math in Python. It introduces the "Array," which is like a standard Python list but exponentially faster and capable of complex mathematical operations.

```python
import numpy as np

# A standard Python list
python_list = [1, 2, 3, 4, 5]

# A NumPy Array
numpy_array = np.array([1, 2, 3, 4, 5])

# Why use NumPy? Math is incredibly easy.
# If we want to multiply every number by 10:
# python_list * 10 (This would just copy the list 10 times, which is bad!)

# NumPy does the math to every element instantly (called "vectorization"):
result_array = numpy_array * 10 
print(f"Multiplied Array: {result_array}")

# Built-in statistical methods
print(f"Average: {np.mean(numpy_array)}")
print(f"Highest Value: {np.max(numpy_array)}")
```

## Lesson 3: Pandas for Data Manipulation
Pandas is built on top of NumPy. It introduces the DataFrame, which is essentially an interactive spreadsheet inside Python. You will use Pandas to load data, clean up missing values, and filter information.

```python
import pandas as pd

# Creating a DataFrame from a dictionary
data = {
    "House_ID": [101, 102, 103, 104, 105],
    "Sq_Ft": [1200, 2500, None, 1800, 3200], # Notice the missing data (None)
    "Price": [150000, 300000, 210000, 250000, 450000]
}

df = pd.DataFrame(data)

# 1. Viewing the data
print("Raw Data:")
display(df) # 'display' makes DataFrames look pretty in Jupyter

# 2. Cleaning Data: Fill the missing 'Sq_Ft' value with the average of the column
average_sq_ft = df["Sq_Ft"].mean()

# Assign the filled data back to the column directly
df["Sq_Ft"] = df["Sq_Ft"].fillna(average_sq_ft)

# 3. Filtering Data: Find houses over $200,000
expensive_houses = df[df["Price"] > 200000]
print("\nHouses over $200k:")
display(expensive_houses)
```

## Lesson 4: Matplotlib & Seaborn for Visualization
Humans are visual learners. We use Matplotlib (the base plotting library) and Seaborn (a library built on top of Matplotlib that makes charts look beautiful with less code) to spot trends in our data.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Using the cleaned DataFrame from Lesson 3

# Set the style of the plots
sns.set_theme(style="whitegrid")

# Create a scatter plot to see the relationship between size and price
plt.figure(figsize=(8, 5)) # Set the size of the chart
sns.scatterplot(data=df, x="Sq_Ft", y="Price", s=100, color="blue")

# Add titles and labels
plt.title("House Price vs. Square Footage", fontsize=16)
plt.xlabel("Square Footage", fontsize=12)
plt.ylabel("Price ($)", fontsize=12)

# Show the plot
plt.show()
```
